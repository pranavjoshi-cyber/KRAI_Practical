```python
#Question-4 Write a multiple regression program for predicting CO2 emission level when Volume
# 1300 of Car is and Engine weight is 3300Kg. Refer Car data file. 

import pandas as pd
import numpy as np
import statsmodels.api as sm
data = pd.read_excel("C:/Users/Pranav/ML DL practical/Datasets/Car_data.xlsx")
```

```python
data
```

```python
X = data[['Volume', 'Weight']]
y = data['CO2']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Predict CO2 emissions for a car with Volume = 1300 and Weight = 3300 Kg
volume = 1300
weight = 3300
predicted_co2 = model.predict([1, volume, weight])

print(f"Predicted CO2 Emissions: {predicted_co2[0]:.2f} g/km")
```
