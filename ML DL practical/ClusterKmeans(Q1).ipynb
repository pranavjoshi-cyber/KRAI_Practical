```python
# #Question-1 Write a program to cluster a set of points using K-means. Consider, K=4, clusters.
# Consider Euclidean distance as the distance measure. Randomly initialize a cluster
# mean as one of the data points. Iterate for 10 iterations. After iterations are over, print
# the final cluster means for each of the clusters.
# Data set iris data set

import numpy as np
import pandas as pd
data = pd.read_csv("C:/Users/Pranav/ML DL practical/Datasets/iris.csv")
```

```python
#1)
# Extract the feature columns (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm)
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values

# Set the number of clusters (K) and maximum iterations
K = 4
max_iterations = 10

# Randomly initialize cluster means as data points
np.random.seed(0)
initial_means_indices = np.random.choice(X.shape[0], K, replace=False)
cluster_means = X[initial_means_indices]

# Helper function to calculate Euclidean distance
def euclidean_distance(point1, point2):
    return np.linalg.norm(point1 - point2)

# K-means clustering
for iteration in range(max_iterations):
    # Initialize cluster assignments for each data point
    cluster_assignments = []

    # Assign each data point to the nearest cluster mean
    for point in X:
        distances = [euclidean_distance(point, mean) for mean in cluster_means]
        closest_cluster = np.argmin(distances)
        cluster_assignments.append(closest_cluster)

    # Update cluster means
    new_cluster_means = []
    for cluster_id in range(K):
        cluster_points = X[np.array(cluster_assignments) == cluster_id]
        if len(cluster_points) == 0:
            # Handle empty clusters by not updating their mean
            new_cluster_means.append(cluster_means[cluster_id])
        else:
            new_mean = np.mean(cluster_points, axis=0)
            new_cluster_means.append(new_mean)

    cluster_means = np.array(new_cluster_means)

# Print the final cluster means
for i, mean in enumerate(cluster_means):
    print(f'Cluster {i + 1} Mean: {mean}')
```

```python
#2)
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

# Load the Iris dataset
data = pd.read_csv("C:/Users/Pranav/ML DL practical/Datasets/iris.csv")
X = data.iloc[:, 1:5].values  # Extract the feature columns

# Set the number of clusters (K)
K = 4

# Create a KMeans instance and fit the data
kmeans = KMeans(n_clusters=K, random_state=0, max_iter=10)
kmeans.fit(X)

# Get the cluster means
cluster_means = kmeans.cluster_centers_

# Print the final cluster means
for i, mean in enumerate(cluster_means):
    print(f'Cluster {i + 1} Mean: {mean}')
```
