```python
#Question-2 The table below gives the amount of Crabby Cakes made by Bakers for each year heâ€™s
# worked.
# Graph the data on a scatter plot, find the line of best fit, and write the equation for the
# line you draw.
# Years worked 1 2 3 4 5 6
# Cakes made 6,500 7,805 10,835 11,230 15,870 16,387
# Correlation Coefficient (r): _________
# Type of Correlation: ______________________
# Using the linear regression equation predict how many Crabby cakes he will make
# after working 10 years. 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

years_worked = np.array([1, 2, 3, 4, 5, 6])
cakes_made = np.array([6500, 7805, 10835, 11230, 15870, 16387])

# Calculate the line of best fit
slope, intercept, r_value, p_value, std_err = linregress(years_worked, cakes_made)

# Calculate the predicted values for the regression line
predicted_cakes = slope * years_worked + intercept

# Plot the data points and the regression line
plt.scatter(years_worked, cakes_made, label='Data Points')
plt.plot(years_worked, predicted_cakes, label='Regression Line', color='red')
plt.xlabel('Years Worked')
plt.ylabel('Cakes Made')
plt.legend()
plt.show()

# Print the correlation coefficient (r)
print("Correlation Coefficient (r):", r_value)
```

```python
years_to_predict = 10
predicted_cakes_10_years = slope * years_to_predict + intercept
print(f"Predicted Cakes after 10 Years: {predicted_cakes_10_years}")

#print("Predicted Cakes after 10 Years:",predicted_cakes_10_years)

```
