```python
#question 11 Write a program to classify the social network Advertise data2 using Bay’s
# classification method
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
data = pd.read_csv("C:/Users/Pranav/ML DL practical/Datasets/Social_Network_Ads2.csv")
```

```python
data
```

```python
X = data.iloc[:, 2:4].values  # Using Age and EstimatedSalary as features
y = data["Purchased"].values

# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Initialize and train the Naive Bayes classifier
classifier = GaussianNB()
classifier.fit(X_train, y_train)

# Predict the target values for the test set
y_pred = classifier.predict(X_test)

# Evaluate the classifier
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```
