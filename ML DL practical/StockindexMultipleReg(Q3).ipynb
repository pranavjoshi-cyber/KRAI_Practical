```python
#3) question-3 Write a program for multiple regression for predicting stock_index_price (dependent
# Variable) using two independent variables (interest_rate and unemployment_rate).
# Refer Stock_data file
# Predict Stock index price for interest rate = 3 and unemployment rate is 5.7


import numpy as np

import pandas as pd
import statsmodels.api as sm

data = pd.read_excel("E:/Imcc lms/Sem3/KRAI Practical/Datasets/Stock_data.xlsx")
```

```python
data
```

```python

# Select the independent variables (X) and the dependent variable (y)
X = data[['Interest Rate', 'Unemployement_rate']]
y = data['Stock_index_price']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Predict stock index price for interest rate = 3 and unemployment rate = 5.7
interest_rate = 3
unemployment_rate = 5.7
predicted_stock_price = model.predict([1, interest_rate, unemployment_rate])

print(f"Predicted Stock Index Price: {predicted_stock_price[0]:.2f}")
```
